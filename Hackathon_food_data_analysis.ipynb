```python
import sqlite3

#connect to database
conn=sqlite3.connect("restaurants.db")

#load and execute sql file
with open("restaurants.sql","r") as f:
  sql_script=f.read()
  # Add DROP TABLE IF EXISTS statement to ensure a clean slate
  sql_script_with_drop = "DROP TABLE IF EXISTS restaurants;" + sql_script
  conn.executescript(sql_script_with_drop)
  conn.commit()

  print("Restaurant table created successfully")
```

```python
restaurants = pd.read_sql(
    "SELECT * FROM restaurants",
    conn
)

restaurants.head()
```

```python
orders_users = orders.merge(
    users,
    on="user_id",
    how="left"
)
```

```python
final_df = orders_users.merge(
    restaurants,
    on="restaurant_id",
    how="left"
)
```

```python
final_df.to_csv(
    "final_food_delivery_dataset.csv",
    index=False
)
```
